```python
import pandas as pd
import numpy as np

# -----------------------------
# 1. Load OGLE-III photometry
# -----------------------------
o3data = pd.read_csv('ogle3_full.dat', delim_whitespace=True, header=None, usecols=range(11))
o3data.columns = ['id', 'i', 'v', 'period', 'period_uncertainty', 't_max', 'amplitude', 'R_21', 'phi_21', 'R_31', 'phi_31']

# Convert missing values to NaN
o3data.replace({'-': np.nan, '99.999': np.nan}, inplace=True)

numeric_cols = ['i', 'v', 'period', 'amplitude']
o3data[numeric_cols] = o3data[numeric_cols].astype(float)
```

```python
# -----------------------------
# 2. Compute Wesenheit index
# -----------------------------
# W_I = I - 1.55*(V - I)
o3data['i'] = o3data['i'].astype(float)
o3data['v'] = o3data['v'].astype(float)
o3data['windex'] = o3data['i'] - 1.55 * (o3data['v'] - o3data['i'])
```

```python
# -----------------------------
# 3. Load ident data
# -----------------------------
ident = pd.read_csv('ogle3_ident_full.dat', delim_whitespace=True, header=None, usecols=[0,3])
ident.columns = ['id','mode']

# Strip whitespace
ident['mode'] = ident['mode'].str.strip()
```

```python
# Ensure 'id' columns are strings
o3data['id'] = o3data['id'].astype(str).str.strip()
ident['id'] = ident['id'].astype(str).str.strip()

# Merge on the 'id' column
merged = pd.merge(o3data, ident, on='id', how='inner')

# Extract type from ID
merged['type'] = merged['id'].str.split('-').str[2]

# Keep only the desired columns
cols = ['id', 'type', 'mode', 'i', 'v', 'period', 'amplitude', 'windex']
merged = merged[cols]
```

```python
# -----------------------------
# 5. Save merged CSV
# -----------------------------
merged_file = "ogle3_full_with_windex.csv"
merged.to_csv(merged_file, index=False)
print(f"Saved merged dataset with Wesenheit index and mode to '{merged_file}'")
```

```python
import os
import pandas as pd
from glob import glob
```

```python
# -----------------------------
# CONFIG
# -----------------------------
dataset_dir = "datasets"
image_dir = "lightcurve_plots"
```

```python
# -----------------------------
# 1. Collect all IDs from datasets
# -----------------------------
dataset_files = glob(os.path.join(dataset_dir, "*.csv"))

all_ids = set()

print("Scanning dataset files...")

for file in dataset_files:
    df = pd.read_csv(file)
    
    if 'id' not in df.columns:
        print(f"[WARNING] 'id' column missing in {file}")
        continue
    
    ids = df['id'].astype(str).tolist()
    all_ids.update(ids)

print(f"Total unique IDs from datasets: {len(all_ids)}")
```

```python
# -----------------------------
# 2. Collect all image filenames
# -----------------------------
image_files = glob(os.path.join(image_dir, "*.png"))

image_ids = set(os.path.splitext(os.path.basename(f))[0] for f in image_files)

print(f"Total images found: {len(image_ids)}")
```

```python
# -----------------------------
# 3. Find missing images
# -----------------------------
missing_images = all_ids - image_ids

if missing_images:
    print(f"\n[ERROR] Missing images: {len(missing_images)}")
    
    # Show a few examples
    for i, mid in enumerate(list(missing_images)[:10]):
        print(f"  Missing: {mid}.png")
else:
    print("\nAll dataset entries have corresponding images")
```

```python
# -----------------------------
# 4. Find orphan images
# -----------------------------
orphan_images = image_ids - all_ids

print(f"\nOrphan images (to delete): {len(orphan_images)}")
```

```python
# -----------------------------
# 5. Delete orphan images
# -----------------------------
for img_id in orphan_images:
    path = os.path.join(image_dir, f"{img_id}.png")
    try:
        os.remove(path)
    except Exception as e:
        print(f"Failed to delete {path}: {e}")

print("\nCleanup complete 🧹")
```
